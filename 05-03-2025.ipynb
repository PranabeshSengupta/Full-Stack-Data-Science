```python
x=[10,20,30]
y1=[22,33,44]
y2=[55,66,77]
#plt.figure(figsize=[4,4])
plt.plot(x,y1,c='g',ms=15,mec='b',mew=4,lw=5,mfc='y', marker='o',label='data-1')
plt.plot(x,y2,c='b',ms=15,mec='b',mew=4,lw=5,mfc='y',marker='+',label='data-2')
plt.legend()
#plt.grid()
plt.axis('on')
plt.title('year wise rainfall')
plt.xlabel('year')
plt.ylabel('rainfall')
```
