```python
a=np.random.randint(10,100,size=(7,5))
a
```

```python
df2=df.mask(df>80)  #fill up missing value 1.fill(fillna()) or 2.remove missing value(dropna())
df2                 #we can use it both dataframe or series
```

```python
df2.fillna(100)
```

```python
df2.fillna(method='ffill')  #it shows warning for value raplecation
```

```python
df2
```

```python
df2.ffill()                 # it is used for forward fillup
```

```python
df2.bfill()                # it is used for backward fillup
```

```python
df2.bfill(axis=1)  #row wise fill in backward direction
```

```python
df2.ffill(axis=1) #row wise fill in forward direction
```
