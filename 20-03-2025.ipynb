```python
df.drop(columns=['RowNumber',	'CustomerId'	,'Surname'],inplace=True)
```

```python
from sklearn.preprocessing import OrdinalEncoder,StandardScaler
```

```python
ord=OrdinalEncoder()

df[['Geography'	,'Gender']]=ord.fit_transform(df[['Geography'	,'Gender']])
df
```

```python
st=StandardScaler()
df[['Balance'	,	'EstimatedSalary']]=st.fit_transform(df[['Balance'	,	'EstimatedSalary']])
df
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(criterion='entropy')
dt.fit(X,y)
```

```python
dt.predict([[619,	0.0,	0.0,	42,	2,	-1.225848	,1,	1,	1	,0.021886]])
```

```python
from sklearn.pipeline import Pipeline
```

```python

p1=Pipeline(
    [
   
        ('st1',OrdinalEncoder()),
          ('st2',StandardScaler()),

]
        )

c1=ColumnTransformer([
    ('drp','drop',[0,1,2]),
    ('tr1',p1,[4])
],remainder='passthrough')
c1
```

```python
pd.DataFrame(c1.fit_transform(df),columns=c1.get_feature_names_out())
```

```python
c1=ColumnTransformer([
    ('drp','drop',[0,1,2]),
    ('enc',OrdinalEncoder(),[4,5]),
    ('scale',StandardScaler(),['Balance'	,	'EstimatedSalary'	])
    
],remainder='passthrough')
c1
```

```python
p1=Pipeline([

    ('st1',c1),
    ('st2',DecisionTreeClassifier(criterion='entropy'))

])

p1
```
