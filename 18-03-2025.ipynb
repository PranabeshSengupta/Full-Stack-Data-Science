```python
max_sc=0
for i in [3,5,7]:
    for j in [1,2]:
        for k in ['uniform','distance']:
                model=KNeighborsClassifier(n_neighbors=i,p=j,weights=k)
                model.fit(X_train,y_train)
                sc=model.score(X_test,y_test)
                print(f'n_neighbors : {i}  p={j}  weights={k}  sc={sc}')
                if sc>max_sc:
                    max_sc=sc
                    
                    best_comb={'n_neighbors':i,'p':j,'weight':k}


print('=================')
best_comb
                    
```
