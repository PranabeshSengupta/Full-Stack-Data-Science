## chatbot application

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
import json
```

```python
with open('intents.json') as f:
    r=json.load(f)

r
```

```python
doc=r['intents']
doc
```

```python
classes=[]
patterns=[]
for i in doc:
    if i['patterns']:
        ##it takes the pattern
        print(i)
        classes.append(i['tag'])
        patterns.extend(i['patterns'])
        print('====')
```

```python
classes
```

```python
len(classes)
```

```python
patterns
```

```python
#vocabulary
```

```python
cv=CountVectorizer(stop_words='english')
```

```python
cv
```

```python
cv.fit_transform(patterns)
```

```python
d=cv.fit_transform(patterns).toarray()
d
```

```python
cv.get_feature_names_out()
```

```python
df=pd.DataFrame(d,columns=cv.get_feature_names_out())
df
```

```python
y=[]
for i in patterns:
    for j in doc:
        if i in j['patterns']:
            y.append(j['tag'])
y
```

```python
len(y)
```

## model creation

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense,Dropout
```

```python
X=df
```

```python
X.shape
```

```python
model = Sequential([Dense(100,activation='relu',input_dim=X.shape[1]),Dense(9,activation='Softmax')])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
en=LabelEncoder()
y_encoded=en.fit_transform(y)
y_encoded
```

```python
model.fit(X,y_encoded,epochs=100,validation_split=.2)
```
