```python
d={'st_name':['ajay','raja','ajay','raja','raja'],
   'subject':['phy','phy','math','math','chem'],
   'marks':[80,99,60,88,65]
  }
df=pd.DataFrame(d)
```

```python
df
```

```python
pv=df.pivot(columns='subject',index='st_name',values='marks')
pv
```

```python
import seaborn as sns
```

```python
sns.heatmap(pv)
```

```python
df2=df[df.country.isin(['India','Pakistan','Sweden'])]
df2
```
