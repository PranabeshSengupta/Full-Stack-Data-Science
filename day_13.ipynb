```python
df=pd.read_csv('cars.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder,OneHotEncoder
```

```python
x=np.array(['mango','banana','apple'])
lb=LabelEncoder()
ord=OrdinalEncoder()
ohe=OneHotEncoder(sparse_output=False)
```

```python
lb.fit_transform(x) #1d array passing...
```

```python
ord.fit_transform(x.reshape(-1,1)) #2d array passing...
```

```python
ohe.fit_transform(x.reshape(-1,1)) #arrange item in alphabetic
```
