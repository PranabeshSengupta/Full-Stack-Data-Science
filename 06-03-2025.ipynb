```python
x=np.array([10,20,30,40,100,200])
plt.boxplot(x)
```

```python
q1=np.percentile(x,25)
q3=np.percentile(x,75)
iqr=q3-q1


```

```python
min=q1-1.5*iqr
max=q3+1.5*iqr
```

```python
p=x[ (x>min) & (x<max) ]
p
```

```python
plt.boxplot(x[ (x>min) & (x<max) ])
```

```python
q1=np.percentile(p,25)
q3=np.percentile(p,75)
iqr=q3-q1

min=q1-1.5*iqr
max=q3+1.5*iqr
```

```python
p2=p[(p>min) & (p<max) ]
plt.boxplot(p2)
```

```python
x=np.array([1,2,3,4,5,100,200])
q1=np.percentile(x,25)
q3=np.percentile(x,75)
iqr=q3-q1
min=q1-iqr*1.5
max=q3+iqr*1.5
c=(x>min) & (x<max)
p1=x[c]
p1
```

```python
plt.boxplot(p1)
```
