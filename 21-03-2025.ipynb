```python
from keras.models import Sequential
```

```python
from keras.layers import Dense,Flatten,Conv2D,MaxPool2D,Rescaling
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('j:/data_science/csv/classification/Cancer_data_1500_V2.csv')
df
```

```python
model=Sequential([
    Dense(10,input_dim=8,activation='sigmoid'),
    Dense(1,activation='sigmoid')
    

])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train, y_test=train_test_split(X,y)
model.fit(X_train,y_train,epochs=10)
```

```python
X_train.shape
```

```python
p=model.predict(X_test)
p
```

```python
pred=[ 1 if i>.5 else 0  for i in p]
pred
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,accuracy_score
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,pred)
```

```python
accuracy_score(y_test,pred)
```
