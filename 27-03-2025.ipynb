```python
from sklearn.model_selection import train_test_split
```

```python
X=df.Message
y=df.Category
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
ct=CountVectorizer(stop_words='english',token_pattern=r'\b[a-zA-Z]{2,}\b')
```

```python
ct
```

```python
ct.fit_transform(X_train)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
ct.vocabulary_
```

```python
d=pd.DataFrame(ct.fit_transform(X_train).toarray(),columns=ct.get_feature_names_out())
d
```

```python
X_train_Coded=ct.fit_transform(X_train)
X_train_Coded
```

```python
X_test_Coded=ct.transform(X_test)
X_test_Coded
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
```

```python
model.fit(X_train_Coded,y_train)
```

```python
model.predict(X_test_Coded)
```

```python
model.score(X_test_Coded,y_test)
```

```python
from sklearn.pipeline import Pipeline
```

```python
p1=Pipeline(steps=[

    ('st1',CountVectorizer(stop_words='english',token_pattern=r'[a-zA-Z]{2,}')),
 ('st2',RandomForestClassifier()),


    ] )
p1
```

```python
p1.fit(X_train,y_train)
```

```python
p1.predict(X_test)
```

```python
p1.score(X_test,y_test)
```
